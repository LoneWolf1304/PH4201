```julia
using Plots, LaTeXStrings, DelimitedFiles, LsqFit
pgfplotsx()
```

```julia
A = readdlm("data.dat")
```

```julia
lin(x,p) = @. x*p[1] + p[2]
fiit = curve_fit(lin, A[:,1], A[:,2], [2.1,0.1])
m = round(fiit.param[1], digits=3)
err = round(standard_errors(fiit)[1], digits=3)
```

```julia
scatter(A[:,1], A[:,2], msw=0.8, label="Data Points", xlabel=L"\beta\ (\mathrm{rad})", ylabel=L"Phase shift $\Delta \Phi$")
plot!(A[:,1], lin(A[:,1], fiit.param), label=L"Linear Fit, slope = $ %$m \pm %$err $", legend=:topleft, size=(450,400))
```
