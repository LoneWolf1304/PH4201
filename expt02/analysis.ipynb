```julia
using Plots, LaTeXStrings, DelimitedFiles, Printf, LsqFit
```

```julia
plotlyjs()
```

```julia
A = readdlm("data1.dat")
x1 = A[:,3]
y1 = A[:,4]
lin(x,p) =@. p[1]*x+p[2]
fiit = curve_fit(lin, y1,x1, [0.1,0.1])
println(fiit.param)
println(standard_errors(fiit))
```

```julia
plot(y1, lin(y1, fiit.param),lw=2, label="Linear fit, slope = 2.045 ± 0.011 ", guidefontsize=20, legendfontsize=11, tickfontsize=20, size=(500,450))
scatter!(y1,x1, label="Data", legend=:top, c="red")
plot!(bottom_margin=5Plots.mm, xticks=1:9)
ylabel!("χ²")
xlabel!("m")

```

```julia
savefig("distance.pdf")
```

```julia
B = readdlm("data2.dat")
x2 = B[:,5]
y2 = B[:,1]
lin(x,p) =@. p[1]*x+p[2]
fiit = curve_fit(lin, y2,x2, [0.1,0.1])
println(fiit.param)
println(standard_errors(fiit))
```

```julia
plot(y2, lin(y2, fiit.param),lw=2, label="Linear fit, slope = 1.656 ± 0.012 ", guidefontsize=20, legendfontsize=11, tickfontsize=20, size=(500,450))
scatter!(y2,x2, label="Data", legend=:top, c="red")
plot!(bottom_margin=5Plots.mm, xticks=1:9)
ylabel!("χ²")
p1 = xlabel!("m")

```
