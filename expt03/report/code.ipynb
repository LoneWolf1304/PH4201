```julia
using Plots, LaTeXStrings

H_1 = [3.9,5.1,3.6,1.6]
P_1 = [16.8,40.8,2.4,51.4]

H_2 = [0.8, 2.4, 2.5, 1.3]
P_2 = [16.6,19.1,17, 32.2]

H_3 = [0.4, 4, 2.7,1.5]
P_3 = [12.3,39.2,2.6,50]

H_4 = [0.5, 3.7,2.7,1.4]
P_4 =[15.3, 39.6, 15.3, 50.5]

H_1 = H_1/sum(H_1)
H_2 = H_2/sum(H_2)
H_3 = H_3/sum(H_3)  
H_4 = H_4/sum(H_4)

P_1 = P_1/sum(P_1)
P_2 = P_2/sum(P_2)
P_3 = P_3/sum(P_3)
P_4 = P_4/sum(P_4)

X_1 = H_1[2] + P_1[3]
Y_1 = H_1[3] - P_1[2]

X_2 = H_2[2] + P_2[3]
Y_2 = H_2[3] - P_2[2]

X_3 = H_3[2] + P_3[3]
Y_3 = H_3[3] - P_3[2]

X_4 = H_4[2] + P_4[3]
Y_4 = H_4[3] - P_4[2]

X = [X_1, X_2, X_3,X_4]
Y = [Y_1, Y_2, Y_3,Y_4]

delta(x1,y1) = acos(sqrt(x1^2+y1^2)-1)
psi(x1, y1)= 0.5 * atan(y1/x1)

deltarr = [delta(X[i],Y[i]) for i in 1:(length(X))]
psarr = [psi(X[i],Y[i]) for i in 1:(length(X))]

A = [0,0.25,0.5,1]
plot(A, psarr, marker=:circle, c="red", label="")
xlabel!("Grayscale Value", ylabelfontsize=18, xlabelfontsize=13, tickfontsize=17)
p1 = ylabel!(L"$\psi $")

plot(A, deltarr, marker=:circle, c="red", label="",ylabelfontsize=18, xlabelfontsize=13, tickfontsize=17)
xlabel!("Grayscale Value")
p2=ylabel!(L"$\delta$")
```

```julia
plot(p1,p2, size=(1000,350), bottom_margin=5Plots.mm, left_margin=5Plots.mm)
```

```julia
using Plots, LaTeXStrings
pgfplotsx()
H_1 = [3.9,5.1,3.6,1.6]
P_1 = [16.8,40.8,2.4,51.4]

H_2 = [0.8, 2.4, 2.5, 1.3]
P_2 = [16.6,19.1,17, 32.2]

H_3 = [0.4, 4, 2.7,1.5]
P_3 = [12.3,39.2,2.6,50]

H_4 = [0.5, 3.7,2.7,1.4]
P_4 =[15.3, 39.6, 15.3, 50.5]

function X(H,P)
    S_1 = [H[1]+H[2],H[1]-H[2],H[3]-H[4]]./(H[1] + H[2])  # Stokes Vector S_H
    S_2 = [P[1]+P[2],P[1]-P[2],P[3]-P[4]]./(P[1] + P[2])  # Stokes Vector S_P
    X = S_1[2] + S_2[3] 
    Y = S_1[3] - S_2[2]
    return X,Y,S_1,S_2
end

X_1,Y_1,SH_1,SP_1 = X(H_1,P_1)
X_2,Y_2,SH_2,SP_2 = X(H_2,P_2)
X_3,Y_3,SH_3,SP_3 = X(H_3,P_3)
X_4,Y_4,SH_4,SP_4 = X(H_4,P_4)

Xarr = [X_1, X_2, X_3, X_4]
Yarr = [Y_1, Y_2, Y_3, Y_4]

delta(x1,y1)= acos(sqrt(x1^2 + y1^2)-1)
psi(x1, y1) = 0.5 * atan(y1,x1)

deltarr = [delta(Xarr[i],Yarr[i]) for i =1:(length(Xarr))]
psarr = [psi(Xarr[i],Yarr[i]) for i in 1:(length(Xarr))]
A = [0,0.25,0.5,1]
```

```julia
plot(A, psarr, marker=:circle, c="red", ms=6, msw=2, lw=2, label="",ylabelfontsize=18, xlabelfontsize=13, tickfontsize=17)
xlabel!("Grayscale Value")
p1=ylabel!(L"$\psi$")
```

```julia
plot(A, deltarr, marker=:circle, c="red", ms=6, msw=2, lw=2, label="",ylabelfontsize=18, xlabelfontsize=13, tickfontsize=17)
xlabel!("Grayscale Value")
p2=ylabel!(L"$\delta$")
```

```julia
plot(p1,p2, size=(1000,350), bottom_margin=5Plots.mm, left_margin=5Plots.mm)
```

```julia
deltI = 0.1
```

```julia
function errors(H_1)
    I = H_1[1]+H_1[2]
    SH2 = 2*deltI/(I)
    SH3 = (2*deltI/(I)) + 2*(H_1[3]-H_1[4])/(I^2)*deltI
    return [SH2,SH3]
end
deltX(H,P) = errors(H)[1]+errors(P)[2]
deltY(H,P) = errors(H)[2]+errors(P)[1]
deltPsi(H,P,X,Y) = (abs(Y)*deltX(H,P)+abs(X)*deltY(H,P))/(2*(X^2+Y^2))
deltDelta(H,P,X,Y) = (abs(X)*deltX(H,P)+abs(Y)*deltY(H,P))/(sqrt(X^2+Y^2)*sqrt(1 - (sqrt(X^2+Y^2)-1)^2))
```
